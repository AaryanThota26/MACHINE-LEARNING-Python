## MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
# it is the visualisation library 
import matplotlib.pyplot as plt
# if we use this below fn then no need to type plt.show() the below fn is supported by thr jupyter
%matplotlib inline
```

```python
import numpy as np
```

```python
## simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## ploting using Matplotlib 

# plt(plot) scatter x is x-axis y is y-axis and c is colour
plt.scatter(x,y,c='r')
# labeling the x axis and y axis 
plt.xlabel('X axis')
plt.ylabel('Y axis')
# setting title for the graph
plt.title('Graph in 2-D')
# displaying the graph
plt.show()
# if we want to save the graph then
plt.savefig('2-D.png')
# the below figure is a scatter diagram means just points will be there
```

```python
y=x*x
```

```python
# plt plot      # r is color     # o is mark style   # -- is the representation of the line style like dotted or solid etc 
#(x-axis,y-axis,colour and the mark style like dot,star(🌟),cross + etcc after it determining the line like dotted line or solid line )
plt.plot(x,y,'ro--')
plt.show()
```

```python
# instead of determing the line style after the mark style the we use line style
plt.plot(x,y,'r*',linestyle='solid',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
plt.title('2-D diagram')
```

```python
## Creating Subplots
# in subplot(no.of rows,no.of columns,index)
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.show()
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.show()
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.show()
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  

# Show the figure. 
plt.show()


# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

## Bar Graphs

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Histograms

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  
plt.show()
```

## Pie Chart


```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.5, 0, 0, 0)  # explode 1st slice move the first slice some away

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
# autopct='%1.1f%%' it will display the floating point up to one decimal only 
autopct='%.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
