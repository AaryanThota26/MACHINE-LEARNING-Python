```python
import seaborn as sns
```

## Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr(numeric_only=True)
```

```python
df.dtypes
```

```python
 sns.heatmap(df.corr(numeric_only=True))
```

## JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
# in the kind it is hexagon so the graph will be represented with hexagons
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
```

```python
# in this the kind is regression 
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
# to take the count of the smokers
df['smoker'].value_counts()
```

```python
# hue is the category where the graps are based on that data 
sns.pairplot(df,hue='smoker')
```

## Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
# based on the x axis the y axis is taken by the values of the kde values if the kde is of then the line in the middle of the graph will not be displayed
sns.distplot(df['tip'])
```

```python
# if the kde(kernal density estimation) is false and if we set bins value to the 10 so the y-axis side there will be 10 size of the indexes
sns.distplot(df['tip'],kde=False,bins=10)
```

##  Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about
- box plot
- violin plot
- count plot
- bar plot

```python
## count plot
# when we use this palette={'Male': 'skyblue', 'Female': 'lightpink'} we can change the colours of the bars
# it is taken for the x axis
# in this we need to give only one axis because on the other axis the count will be displayed
sns.countplot(x='sex', data=df,palette='rainbow')
```

```python
# this will take for the y axis
sns.countplot(y='sex',data=df)
```

```python
# Bar plot 
# in bar plot we have to mention both the x axis and the y axis
sns.barplot(y='total_bill',x='smoker',data=df)
```

## Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='sex',y='total_bill', data=df)
```

```python
# we can also give different colours
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
# orient means the side the graph should be displayed like horizontal (h) or vertical (v)
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
